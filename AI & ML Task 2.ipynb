# Feature Engineering

```python
#Adding a grade column to the DataFrame based on the exam_score column
#Building a function to assign appropriate grades to the marks 
def assign_grades(exam_score):
    if(exam_score>=90 and exam_score<=100):
        return "O (Outstanding)"
    elif(exam_score>=80 and exam_score<90):
        return "D+ (Excellent)"
    elif(exam_score>=75 and exam_score<80):
        return "D (Distinction)"
    elif(exam_score>=70 and exam_score<75):
        return "A+ (Very Good)"
    elif(exam_score>=60 and exam_score<70):
        return "A (Good)"
    elif(exam_score>=50 and exam_score<60):
        return "B (Average)"
    else:
        return "U (Re-Appear)"

#Applying the function
df['grade']=df['exam_score'].apply(assign_grades)
```

# Drawing inferences by asking questions from the data
1. How does the time spent on social media and Netflix affect the sleep and mental health of a student ?
2. How many students are working part-time based on gender ?
3. What is the optimal study hours followed by toppers ?
4. Does diet have any impact on the exam scores of a student ?

```python
#1. How does the time spent on social media and Netflix affect the sleep and mental health of a student ?
#Answering the 1st question using a Relational plot (Scatter plot)
ax=sns.relplot(data=df,x=df['social_media_hours']+df['netflix_hours'],y='sleep_hours',hue='mental_health_rating')
ax.set_axis_labels('Total Screen Time','Sleep Hours')
plt.title("Impact of Screen Time on Sleep and Mental Health")
plt.show()
```

```python
#2.  How many students are working part-time based on gender ?
#Answering the 2nd question using a Bar Plot
df[df["part_time_job"] == "Yes"]["gender"].value_counts() \
            .plot(kind='bar',xlabel='Gender',ylabel='No. of Students',color=["lightblue","pink","lavender"])
plt.title("Count of Students working Part-time based on Gender")
plt.show()
```

```python
#3. What is the optimal study hours followed by toppers ?
#Answering the 3rd question using a Relational plot (Scatter plot)
toppers=df[df['exam_score']>=90]
toppers
```

```python
sns.relplot(data=toppers,x='exam_score',y='study_hours_per_day')
plt.title("Optimal Study Hours of Toppers")
plt.show()
```

```python
#4. Does diet have any impact on the exam scores of a students ?
#Answering the 4th question using a horizontal bar plot
df.groupby('diet_quality')['exam_score'].mean() \
        .plot(kind='barh',xlabel="Exam Scores",ylabel="Diet Quality",color=["navy","slateblue","cyan"])
plt.title("Average of scores based on Diet Quality")
plt.show()
```
